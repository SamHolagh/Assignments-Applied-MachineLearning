```python
import pandas as Zoozoo
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as linearegression
from sklearn.metrics import mean_squared_error, r2_score
print("*****************************************************")
print("*                                                   *")
print("*             Sam Rahimzadeh Holagh                 *")
print("*                                                   *")
print("* Linear Regression Assignment - Applied Machine    *")
print("*        Learning for Data Scientists               *")
print("*                                                   *")
print("*****************************************************")





# Dataset
Data = { 'Hours_Studied': [10,8,5,2,1], 'Hours_Sleep' : [2,4,8,10,12], 'Grades' : [4,3,2,1,0]  }
Zizi = Zoozoo.DataFrame(Data)
X= Zizi [['Hours_Studied','Hours_Sleep']]
Y= Zizi [['Grades']]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.4,random_state=42)
model = linearegression()
model.fit(X_train,Y_train)
Y_predict = model.predict(X_test)
mse= mean_squared_error(Y_test,Y_predict)
r2= r2_score(Y_test,Y_predict)
print(mse)
print(r2)

# New data for prediction
new_data = {
    'Hours_Studied': [12],  # Example: 7 hours studied
    'Hours_Sleep': [3]     # Example: 6 hours slept
}

# Convert new data to DataFrame
new_Zizi = Zoozoo.DataFrame(new_data)

# Predict the grade
predicted_grade = model.predict(new_Zizi)

# Display the prediction
print(f"Predicted Grade: {predicted_grade[0]}")
```
