```python
"""
******************************************************************************
*                                                                            *
*                        Sam Rahimzadeh Holagh                               *
*           Applied Machine Learning for Data Scientists (Fall 2024 1)       *
*                                                                            *
******************************************************************************
"""

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score


#Q1 Loading data:

iris = load_iris()
Kit = iris.data
Kat = iris.target

#Q2:Print Shape
print("Shape of the data:", Kit.shape)

#Q3:Print Features
print("Feature names:", iris.feature_names)

#Q4:Print TargetName
print("Target names:", iris.target_names)

#Q5:Split Data
Kittrain, Kittest, Kattrain, Kattest = train_test_split(Kit, Kat, test_size=0.2, random_state=42)

#Q6 and 7: Train and Accuracy Check
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(Kittrain, Kattrain)
Kat_pred = svm_classifier.predict(Kittest)
accuracy = accuracy_score(Kattest, Kat_pred)
print("Accuracy:", accuracy)

#Q8: Classification Rep
print(classification_report(Kattest, Kat_pred, target_names = iris.target_names))

print ("*************************************************************")
print ("**                                                         **")
print ("**                     Sepal Column                        **")
print ("**                                                         **")
print ("*************************************************************")

#Q9:Accuracy based on two sepal columns
Kitsepal = iris.data[:, [0,1]]
Kittrain, Kittest, Kattrain, Kattest = train_test_split(Kitsepal, Kat, test_size=0.2, random_state=42)
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(Kittrain, Kattrain)
Kat_pred = svm_classifier.predict(Kittest)
accuracy = accuracy_score(Kattest, Kat_pred)
print(classification_report(Kattest, Kat_pred, target_names = iris.target_names))

#Extra Play with data!
print ("*************************************************************")
print ("**                                                         **")
print ("**                     Extra Part                          **")
print ("**               Miss calculated portion                   **")
print ("*************************************************************")

for i in range(len(Kattest)):
    predicted_label = iris.target_names[Kat_pred[i]]
    actual_label = iris.target_names[Kattest[i]]
    if predicted_label != actual_label:
        print(f"Misclassified! Predicted: {predicted_label}, Actual: {actual_label}")
    else:
        print(f"Correct: Predicted = Actual = {predicted_label}")

for i in range(len(Kitsepal)):
  print (Kitsepal[i,0],Kitsepal[i,1], iris.target_names[Kat[i]])

  scores = cross_val_score(svm_classifier, Kitsepal, Kat, cv=5)
print("Cross-validation scores for each fold:", scores)
print("Mean cross-validation accuracy:", scores.mean())

print ("*************************************************************")
print ("**                                                         **")
print ("**                     Petal Column                        **")
print ("**                                                         **")
print ("*************************************************************")

#Q10 Accuracy based on two Petal Column
Kitpetal = iris.data[:, [2,3]]
Kittrain, Kittest, Kattrain, Kattest = train_test_split(Kitpetal, Kat, test_size=0.2, random_state=42)
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(Kittrain, Kattrain)
Kat_pred = svm_classifier.predict(Kittest)
accuracy = accuracy_score(Kattest, Kat_pred)
print(classification_report(Kattest, Kat_pred, target_names = iris.target_names))

#implementing cross validation since the scores are too much accurate?????
scores = cross_val_score(svm_classifier, Kitpetal, Kat, cv=5)
print("Cross-validation scores:", scores)
print("Mean cross-validation accuracy:", scores.mean())






```
